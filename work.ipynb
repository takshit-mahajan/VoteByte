```python
import json
```

```python
class ElectionCommission:
    def __init__(self,country,total_states):
        self.country=country
        self.total_states=total_states
    def display_info(self):
        print(f"Country name: {self.country} and total states in country is {self.total_states}")
        
        
```

```python
class StateElectionCommission (ElectionCommission):
    def __init__(self, country, total_states,state_name,num_constituencies ):
        super().__init__(country, total_states)
        self.state_name=state_name
        self.num_constituencies =num_constituencies 
    
    def display_state_info(self):
        self.display_information()
        print(f"State: {self.state_name}")
        print(f"Number of constituencies: {self.num_constituencies}")
    
```

# conversion into bytes object or into binary data:-
1. using json like json converts dict->string then this string is encoded to binary form using .encode('utf-8')


```python
class Candidate:
    def __init__(self,candidate_id:int,name:str,party:str,age:int,is_incumbent:bool,vote_share:float):
        self.candidate_id=candidate_id
        self.name=name
        self.party=party
        self.age=age
        self.is_incumbent=is_incumbent
        self.vote_share=vote_share
        
    def encode_candidate_to_bytes(self):
        candidate_data={
            "candidate_id":self.candidate_id,
            "name":self.name,
            "party":self.party,
            "age":self.age,
            "is_incumbent":self.is_incumbent,
            "vote_share":self.vote_share
            
            
        }
        json_str_format=json.dumps(candidate_data)
        candidate_data_in_bytes=json_str_format.encode('utf-8')
        
        return candidate_data_in_bytes    
    
    @staticmethod
    def decode_candidate_from_bytes(candidate_bytes_data):
        json_str_format=candidate_bytes_data.decode('utf-8')
        data=json.loads(json_str_format)
        return Candidate(
            candidate_id=data["candidate_id"],
            name=data["name"],
            party=data["party"],
            age=data["age"],
            is_incumbent=data["is_incumbent"],
            vote_share=data["vote_share"]
            
        )
              
    
```

```python
# Testing Cell
candidate_one=Candidate(1,"Takshit","BJP",20,False,59.6)

encoded_data = candidate_one.encode_candidate_to_bytes()
print("Encoded Data:")
print(encoded_data)

print()
decoded_candidate = Candidate.decode_candidate_from_bytes(encoded_data)
print("Decoded Candidate Object:")
print(f"ID: {decoded_candidate.candidate_id}")
print(f"Name: {decoded_candidate.name}")
print(f"Party: {decoded_candidate.party}")
print(f"Age: {decoded_candidate.age}")
print(f"Is Incumbent: {decoded_candidate.is_incumbent}")
print(f"Vote Share: {decoded_candidate.vote_share}")
```

```python
def calculate_total_votes(all_votes:bytearray):
    return sum(all_votes)

def update_votes(all_votes:bytearray,index:int,new:int):
    if index<0 or index>len(all_votes):
        return "Not possible"
    all_votes[index]=new
    return all_votes
        
```

```python
votes=bytearray([1,2,3])
print("Initial votes:- ",calculate_total_votes(votes)) 
print()
print("After updating votes:-")
print(update_votes(votes,1,20)) #This is in binary format.
print(calculate_total_votes(votes))
```
